i) $$\begin{equation} \frac{Y(s)}{X(s)} = \frac{125}{s^2+10s+25}\end{equation}$$

#### Transferencia

$$\begin{equation} H(s) = \frac{125}{(s+5)^2}\end{equation}$$

#### Ecuacion diferencial

$$\begin{equation} \frac{d^2y(t)}{dt^2} + 10\frac{dy(t)}{dt} + 25y(t) = 125x(t)\end{equation}$$

#### Polos y ceros

```python
num1 = np.array([0, 0, 125])
den1 = np.array([1, 10, 25])

tf1 = control.tf(num1, den1)
control.pzmap(tf1);
```

#### Respuesta al escalon

```python
t, y = control.step_response(tf1)
#plt.plot(t, y, label= "time");
fig, ax = plt.subplots()
ax.plot(t, y)

ax.set(xlabel='time [s]', ylabel='amplitud []', title='Step Response');
```

$$\begin{equation}y(t) = 5[1-e^{-5t}(1+5t)]u(t)\end{equation}$$

```python
u = np.ones(t.size)
y = 5 * (1 - np.exp(-5 * t) * (1 + 5 * t)) * u

fig, ax = plt.subplots()
ax.plot(t, y)
ax.set(xlabel='time [s]', ylabel='amplitud []', title='Step Response');
```
